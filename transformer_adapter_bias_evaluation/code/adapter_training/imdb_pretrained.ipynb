```python
from transformers import AutoModelWithHeads, AdapterConfig, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

model = AutoModelWithHeads.from_pretrained("bert-base-uncased")
adapter_name = model.load_adapter("sentiment/imdb@ukp", config="pfeiffer")

print(model.config.adapters.adapters)
model.set_active_adapters(adapter_name)

#model.train_adapter("imdb")
```

```python
model.save_adapter(path +"models/imdb-pretrained/", "imdb")
```
