```python
from transformers import AutoModelWithHeads, AdapterConfig, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelWithHeads.from_pretrained("distilbert-base-uncased")
adapter_name = model.load_adapter("sentiment/rotten_tomatoes@ukp", config="pfeiffer")


print(model.config.adapters.adapters)
model.set_active_adapters(adapter_name)

```
