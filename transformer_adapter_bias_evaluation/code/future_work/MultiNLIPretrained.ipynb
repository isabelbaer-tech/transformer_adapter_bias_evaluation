```python
import sys
path = "/content/gdrive/MyDrive/master_hpi/NLP_Project/code/"
sys.path.append(path)
```

```python
from transformers import AutoModelWithHeads, AdapterConfig
model = AutoModelWithHeads.from_pretrained("gpt2")
config = AdapterConfig.load("houlsby", non_linearity="swish", reduction_factor=16)
model.load_adapter("nli/multinli@ukp", config=config)
print(model.config.adapters.adapters)
model.set_active_adapters("mnli")
#model.train_adapter("mnli")
```

```python
model.save_adapter(path +"models/multinlipretrained/", "mnli")
```
